```python
import time
from playwright.async_api import async_playwright
```

```python
pw = await async_playwright().start()
```

```python
browser = await pw.chromium.launch(headless=True, timeout=0)
```

```python
page = await browser.new_page()
```

```python
await page.goto("https://www.tntsupermarket.com/eng/weekly-special-er.html")
```

```python
async def scrollPageDown(page, scrollWaitSeconds, endOfPageSelector):
  selector = None
  while(selector == None):
    await page.keyboard.press("PageDown")
    time.sleep(scrollWaitSeconds)
    selector = await page.query_selector(endOfPageSelector)
    print(selector)
    
async def getItemList(page, itemSelector):
  itemSelectorList = await page.query_selector_all(itemSelector)
  itemList = []
  for itemSelector in itemSelectorList:
    itemList.append(await itemSelector.inner_html())
  return itemList
  
```

```python
await scrollPageDown(page=page,scrollWaitSeconds=1, endOfPageSelector="p[class^='category-noData-']")
```

```python
itemList = await getItemList(page=page, itemSelector="div[class^='item-root-']")
```
